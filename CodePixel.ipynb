##Step 4 — Abrir el archivo Excel

```python
wb = openpyxl.load_workbook(file_path)
ws = wb[wb.sheetnames[0]]  # usa la primera hoja
```

##Step 5 — Función para obtener el RGB del color de cada celda

```python
def get_rgb(cell):
    fill = cell.fill
    if fill and fill.fgColor and fill.fgColor.rgb:
        rgb = fill.fgColor.rgb
        if len(rgb) == 8:
            return rgb[2:]   # quitar canal alpha
        return rgb
    return ""
```

##Step 6 — Crear una nueva hoja con los códigos RGB

```python
output_sheet = wb.create_sheet("RGB_CODES")

for row in ws.iter_rows():
    for cell in row:
        rgb_value = get_rgb(cell)
        new_cell = output_sheet.cell(row=cell.row, column=cell.column)
        new_cell.value = rgb_value
        new_cell.fill = PatternFill(fill_type=None)  # quitar color
```
